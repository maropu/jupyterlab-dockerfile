```python
import sqlite3

conn = sqlite3.connect('schemaspy_test.db')
conn.execute("PRAGMA foreign_keys = 1")

conn.execute("""
CREATE TABLE item (
   item_code INTEGER PRIMARY KEY,
   item_name TEXT NOT NULL UNIQUE
);
""")

conn.execute("INSERT INTO item VALUES(1, 'food');")
conn.execute("INSERT INTO item VALUES(2, 'house');")
conn.execute("INSERT INTO item VALUES(3, 'infra');")
conn.execute("COMMIT;")

conn.execute("""
CREATE TABLE acc_data ( 
    id INTEGER PRIMARY KEY AUTOINCREMENT,
    acc_date DATE NOT NULL,
    item_code INTEGER NOT NULL,
    amount INTEGER,
    FOREIGN KEY(item_code) REFERENCES item(item_code)
);
""")

conn.execute("""
    INSERT INTO acc_data(acc_date, item_code, amount)
    VALUES('2017-3-1', 1, 1000);
""")
conn.execute("COMMIT;")

conn.commit()
conn.close()
```
